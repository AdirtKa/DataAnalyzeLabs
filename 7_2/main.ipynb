```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import textwrap

# Загрузка файла Excel (укажите свой путь к файлу)
df = pd.read_excel('Kl_1.xlsx')

# Просмотр первых строк таблицы
df.head()
```

```python
df_long = df.melt(
    id_vars=df.columns[0],        # первый столбец — идентификатор (Сектор)
    var_name="Год",               # название нового столбца для лет
    value_name="Значение"         # название столбца с данными
)
df_long["Год"] = pd.to_numeric(df_long["Год"], errors="coerce")
df_long
```

```python
def shorten_label(label: str, width: int = 25) -> str:
    """Автоматически сокращает длинные названия"""
    if len(label) <= width:
        return label
    # если длинное — обрезаем до ближайшего пробела и добавляем "…"
    short = textwrap.shorten(label, width=width, placeholder="…")
    return short

df_long["Сектор_кратко"] = df_long["Сектор"].apply(shorten_label)
```

```python
# создаём интерактивный график
fig = px.line(
    df_long,
    x="Год",
    y="Значение",
    color="Сектор_кратко",
    hover_name="Сектор",       # при наведении — полное имя
    markers=True,
    title="Динамика значений по секторам"
)

# оформление
fig.update_layout(
    legend_title_text="Сектор (сокращённо)",
    legend=dict(
        orientation="v",
        yanchor="top",
        y=1,
        xanchor="left",
        x=1.02
    ),
    hovermode="x unified",
    template="plotly_white",
    width=950,
    height=600
)

fig.show()
```
